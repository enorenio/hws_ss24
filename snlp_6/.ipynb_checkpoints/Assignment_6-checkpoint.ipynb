Aside from the standard smoothing operations, another way of managing OOVs is to look into lower order n-grams when we see unseen tokens. In other words, break down the tokens into their constituents, with the idea that eventually (probably in the next level), we'll be able to reconcile the unseen words.There are generally two ways of doing this: Backoff and  Interpolation.
Here, we're talking about specific such implementations: **Katz Backoff** and **Linear Interpolation**.

Answer for 1: 


#### Linear Interpolation

Linear interpolation is a technique that combines the probabilities from different order n-gram models to produce a final probability estimate. The idea is to use a weighted average of the higher-order n-gram probabilities along with lower-order n-gram probabilities to ensure that even if the higher-order n-gram has not been observed often (or at all), the lower-order n-grams can still contribute to the overall probability estimate

For example, in a trigram model, the probability of a word $ w_i $ given its previous two words $ w_{i-2} $ and $ w_{i-1} $ can be interpolated as follows:

$$ 
P(w_i \mid w_{i-2} w_{i-1}) = \lambda_3 P_{ML}(w_i \mid w_{i-2} w_{i-1}) + \lambda_2 P_{ML}(w_i \mid w_{i-1}) + \lambda_1 P_{ML}(w_i) 
$$


####  Katz Backoff

Katz Backoff is a recursive smoothing technique that uses the highest-order n-gram available but backs off to a lower-order n-gram if the higher-order n-gram has zero counts. It applies a discount to the higher-order n-gram counts and redistributes the leftover probability mass to the lower-order n-grams. The process continues until a known n-gram is found or the unigram level is reached

The complete Katz Backoff formula combines these concepts into a recursive definition:

$$
P_{\text{Katz}}(w_n \mid w_{1}^{n-1}) = 
\begin{cases} 
d_{w_{1}^{n}} \cdot P_{\text{ML}}(w_n \mid w_{1}^{n-1}) & \text{if } C(w_{1}^{n}) > 0 \\
\alpha(w_{1}^{n-1}) \cdot P_{\text{Katz}}(w_n \mid w_{2}^{n-1}) & \text{otherwise}
\end{cases}
$$

where $ P_{\text{ML}} $ is the maximum likelihood estimate, and $ d_{w_{1}^{n}} $ is a discount factor for the n-gram $ w_1, w_2, ..., w_n $.

 where $ \alpha(w_{1}^{n-1})$ is the backoff weight, which ensures that the total probability mass is correctly distributed.

The backoff weight $ \alpha(w_{1}^{n-1}) $ is calculated to redistribute the probability mass that was discounted from the observed n-grams. It can be calculated as:

$
\alpha(w_{1}^{n-1}) = \frac{1 - \sum_{\text{observed } w_n} P_{\text{Katz}}(w_n \mid w_{1}^{n-1})}{1 - \sum_{\text{observed } w_n} P_{\text{ML}}(w_n \mid w_{1}^{n-1})}
$


#### Difference Between Linear Interpolation and Katz Backoff

| **Linear Interpolation**  | **Katz Backoff**         |
|---------------------------|--------------------------|
| Combines probabilities from multiple n-gram levels using fixed weights. | Uses higher-order n-grams but backs off to lower-order n-grams when counts are zero. |
| Requires careful tuning of weights. | Applies discounts to higher-order counts and redistributes probability mass. |
| Simultaneously considers all levels of n-grams. | Recursively backs off to lower-order n-grams until a known n-gram is found. |
| Optimal with sufficient data to tune weights effectively. | Optimal in data sparse environments where higher-order n-grams are often missing. |

#### Optimal Use : 
| **Linear Interpolation**  | **Katz Backoff**         |
|---------------------------|--------------------------|
| This method is beneficial when you want to leverage information from all levels of n-grams simultaneously and when there is sufficient data to effectively tune the weights. It works well in situations where the probabilities from various n-gram levels need to be balanced to provide a more robust estimate.| Katz Backoff is particularly useful in cases of data sparsity, where higher-order n-grams are often missing. It is optimal when you need a method that prioritizes higher-order n-grams but can fallback effectively to lower orders without overly depending on the tuning of weights.|

Example Case when to use Backoff instead of Linear :  In a corpus with limited occurrences of trigrams or higher-order n-grams, backoff can help by using bigrams or unigrams to fill in the gaps.

After implementing the interpolated n-gram model, you need to find the optimal values for the smoothing parameter $\alpha$ and the n-gram order using 5-fold cross-validation. We have provided with a selection of values (or search space) to find the optimal one from:

```python
def fold_cross_validation_split(data, n_folds=5):
    """Split data into n_folds for cross-validation."""
    fold_size = len(data) // n_folds
    return [data[i*fold_size:(i+1)*fold_size] for i in range(n_folds)]

def evaluate_model(train_sents, test_sents, alpha, order):
    """Train and evaluate the Interpolated Model and return the perplexity."""
    model = exercise_1.Interpolated_Model(train_sents, test_sents, alpha=alpha, order=order)
    return model.perplexity()

# For each fold - for n grams
def plot_perplexity_per_fold(ngram_fold_perp_list):
    plt.figure(figsize=(8, 6))
    markers = ['o', 's', '^', 'D', 'v', 'p', '*', 'h', 'x', '+']  # Define markers for each n-gram order
    colors = plt.cm.rainbow(np.linspace(0, 1, len(ngram_fold_perp_list)))  # Generate colors for each n-gram order

    for idx, (n, fold_perp_list) in enumerate(ngram_fold_perp_list):
            plt.plot([1,2,3,4,5], fold_perp_list, marker=markers[n-1], color=colors[idx], label=f'n-gram order {n}, folds')

    plt.xlabel('Fold')
    plt.ylabel('Perplexity')
    plt.title('Perplexity across folds for each n-gram order')
    plt.legend()
    plt.grid(True)
    plt.show()


# N gram vs fold avg perpexility
def plot_avg_perplexity(ngram_range, avg_perp_list):
    plt.figure(figsize=(8, 6))
    plt.plot(ngram_range, avg_perp_list, marker='o')
    plt.xlabel('n-gram order')
    plt.ylabel('Average Perplexity')
    plt.title('Average Perplexity vs n-gram order')
    plt.grid(True)
    plt.show()


def find_optimal_ngram_order_with_plots(corpus, alpha, ngram_range, n_folds=5):
    """Find the optimal n-gram order using cross-validation."""
    folds = fold_cross_validation_split(corpus, n_folds)
    fold_perp_list = []
    ngram_fold_perp_list = []
    ngram_avg_perp = []

    for n in ngram_range:
        for i in range(n_folds):
            train_sents = [sent for j, fold in enumerate(folds) if j != i for sent in fold]
            test_sents = folds[i]
            perplexity = evaluate_model(train_sents, test_sents, alpha, n)
            fold_perp_list.append(perplexity)
        ngram_fold_perp_list.append((n,fold_perp_list))
        ngram_avg_perp.append(np.mean(fold_perp_list))
        fold_perp_list=[]

#     # Plot perplexity for each fold
    plot_perplexity_per_fold(ngram_fold_perp_list)

    # Plot average perplexity across folds
    plot_avg_perplexity(ngram_range, ngram_avg_perp)

    optimal_n = ngram_range[np.argmin(ngram_avg_perp)]
    return optimal_n
```

```python
# Plot 1 for finding optimal n-gram length
ngram_range = range(1, 10)
alpha = 0.0001
optimal_ngram_order = find_optimal_ngram_order_with_plots(vocabulary_restricted_train, alpha, ngram_range)
print(f'Optimal n-gram order: {optimal_ngram_order}')
```

```python
def plot_avg_perplexity_alpha(alpha_range, avg_perp_list):
    plt.figure(figsize=(10, 6))
    plt.plot(alpha_range, avg_perp_list, marker='o')
    plt.xlabel('Alpha')
    plt.ylabel('Average Perplexity')
    plt.title('Average Perplexity vs Alpha')
    plt.grid(True)
    plt.show()

def find_optimal_alpha_with_plots(corpus, n, alpha_range, n_folds=5):
    """Find the optimal alpha value using cross-validation and plot results."""
    folds = fold_cross_validation_split(corpus, n_folds)
    alpha_avg_perp = []

    for alpha in alpha_range:
        fold_perp_list = []
        for i in range(n_folds):
            train_sents = [sent for j, fold in enumerate(folds) if j != i for sent in fold]
            test_sents = folds[i]
            perplexity = evaluate_model(train_sents, test_sents, alpha, n)
            fold_perp_list.append(perplexity)
        
        avg_perplexity = np.mean(fold_perp_list)
        alpha_avg_perp.append(avg_perplexity)

    # Plot average perplexity across folds for each alpha value
    plot_avg_perplexity_alpha(alpha_range, alpha_avg_perp)

    optimal_alpha = alpha_range[np.argmin(alpha_avg_perp)]
    return optimal_alpha
```

# Exercise 2: Kneser-Ney Smoothing (4 points)

One of the more popular methods for smoothing language models is Kneser-Ney smoothing. In this exercise, we'll use a toy example to get an intuitive understanding of how it works, and compare it against a simpler form of smoothing: add-$\alpha$ smoothing.

Consider the following notation

- $V$ - LM vocabulary
-  $N (x)$ - count of the n-gram x in the training corpus
- $N_+(\bullet , w) \triangleq |\{u : N (u, w) > 0\}|$ - number of bigram types ending in w
- $N_+(w ,\bullet) \triangleq |\{u : N (w, u) > 0\}|$ - number of bigram types starting with w
- $N_+(\bullet, w ,\bullet) \triangleq |\{(u,v) : N (u, w, u) > 0\}|$ - number of trigram types with w in the middle

This exercise aims to provide a basic understanding of Kneser-Ney Smoothing. Kneser-Ney Smoothing makes use of *continuation counts* of words for lower order n-grams, given as

\begin{equation}
C_{KN} =
\begin{cases}
\text{count}(\bullet) & \text{for highest order} \\
\text{continuation count}(\bullet) & \text{for lower orders}
\end{cases}
\end{equation}

For example, for a trigram model, Kneser-Ney Smoothing is implemented using the following equations:

$$P_{KN}(w_3|w_1, w_2) = \frac{\max\{N(w_1 w_2 w_3)-d,0\}}{N(w_1 w_2)} + \lambda(w_1, w_2)P_{KN}(w_3|w_2)$$

$$P_{KN}(w_3|w_2) = \frac{\max\{N_{+}(\bullet, w_2, w_3)-d,0\}}{N_{+}(\bullet, w_2, \bullet)} + \lambda(w_2)P_{KN}(w_3)$$

\begin{equation}
P_{KN}(w_3) = \begin{cases}
\frac{N_{+}(\bullet, w_3)}{N_{+}(\bullet, \bullet)} & \text{if $w_3 \in$ V} \\
\frac{1}{V} & \text{otherwise}
\end{cases}
\end{equation}

$\lambda$ is used to normalise the discounted probability mass and is given by

$$\lambda(w_1, w_2) = \frac{d}{N(w_1, w_2)} \cdot N_{+}(w_1, w_2, \bullet)$$

$$\lambda(w_2) = \frac{d}{N(w_2)} \cdot N_{+}(w_2, \bullet)$$

### 2.1 - Kneser-Ney Smoothing

Kneser-Ney smoothing adjusts the probability of an n-gram by considering not only the counts of the n-gram itself but also the diversity of contexts in which the words appear. The smoothing involves three main components:

Discounting: Reduces the counts of observed n-grams by subtracting a discouting paramter and reallocates this probability mass to less frequent n-grams by using backo-off.
Back-off: Distributes the probability mass of discounted counts to lower-order n-grams.
Continuation: measure how often a word appears as a continuation of different preceding words. This provides a more accurate representation of a word's likelihood in context, especially for lower order n-grams

1. **Trigram Probability**:
$$ P_{KN}(w_i | w_{i-2}, w_{i-1}) = \frac{\max(c(w_{i-2}, w_{i-1}, w_i) - D, 0)}{c(w_{i-2}, w_{i-1})} + \lambda(w_{i-2}, w_{i-1}) P_{KN}(w_i | w_{i-1}) $$

2. **Bigram Probability**:
$$ P_{KN}(w_i | w_{i-1}) = \frac{\max(c(w_{i-1}, w_i) - D, 0)}{c(w_{i-1})} + \lambda(w_{i-1}) P_{KN}(w_i) $$

3. **Unigram Probability (Continuation Probability)**:
$$ P_{KN}(w_i) = \frac{| \{ w_{i-1} : c(w_{i-1}, w_i) > 0 \} |}{\sum_w | \{ w_{i-1} : c(w_{i-1}, w) > 0 \} |} $$

4. **Back-off Weight for Trigrams**:
$$ \lambda(w_{i-2}, w_{i-1}) = \frac{D \cdot | \{ w_i : c(w_{i-2}, w_{i-1}, w_i) > 0 \} |}{c(w_{i-2}, w_{i-1})} $$

5. **Back-off Weight for Bigrams**:
$$ \lambda(w_{i-1}) = \frac{D \cdot | \{ w_i : c(w_{i-1}, w_i) > 0 \} |}{c(w_{i-1})} $$

#### Example

Consider the following counts:
- $$ c(\text{the cat sat}) = 1 $$
- $$ c(\text{the cat}) = 5 $$
- $$ c(\text{cat sat}) = 2 $$
- $$ c(\text{cat}) = 10 $$
- $$ c(\text{sat}) = 3 $$
- Discount $$ D = 0.75 $$

##### Step-by-Step Calculation

1. **Trigram Probability $$ P_{KN}(\text{sat} | \text{the cat}) $$**:

$$ P_{KN}(\text{sat} | \text{the cat}) = \frac{\max(c(\text{the cat sat}) - D, 0)}{c(\text{the cat})} + \lambda(\text{the cat}) P_{KN}(\text{sat} | \text{cat}) $$

$$ = \frac{\max(1 - 0.75, 0)}{5} + \lambda(\text{the cat}) P_{KN}(\text{sat} | \text{cat}) $$

$$ = \frac{0.25}{5} + \lambda(\text{the cat}) P_{KN}(\text{sat} | \text{cat}) $$

$$ = 0.05 + \lambda(\text{the cat}) P_{KN}(\text{sat} | \text{cat}) $$

2. **Calculate the Back-off Weight $$ \lambda(\text{the cat}) $$**:

$$ \lambda(\text{the cat}) = \frac{0.75 \cdot | \{ \text{sat}, \text{is}, \text{runs} \} |}{5} = \frac{0.75 \cdot 3}{5} = 0.45 $$

3. **Bigram Probability $$ P_{KN}(\text{sat} | \text{cat}) $$**:

$$ P_{KN}(\text{sat} | \text{cat}) = \frac{\max(c(\text{cat sat}) - D, 0)}{c(\text{cat})} + \lambda(\text{cat}) P_{KN}(\text{sat}) $$

$$ = \frac{\max(2 - 0.75, 0)}{10} + \lambda(\text{cat}) P_{KN}(\text{sat}) $$

$$ = \frac{1.25}{10} + \lambda(\text{cat}) P_{KN}(\text{sat}) $$

$$ = 0.125 + \lambda(\text{cat}) P_{KN}(\text{sat}) $$

4. **Calculate the Back-off Weight $$ \lambda(\text{cat}) $$**:

$$ \lambda(\text{cat}) = \frac{0.75 \cdot | \{ \text{sat}, \text{sleeps}, \text{jumps}, \ldots \} |}{10} = \frac{0.75 \cdot 7}{10} = 0.525 $$

5. **Unigram Probability $$ P_{KN}(\text{sat}) $$**:

$$ P_{KN}(\text{sat}) = \frac{| \{ \text{cat}, \text{dog}, \text{bird} \} |}{\sum_w | \{ w_{i-1} : c(w_{i-1} w) > 0 \} |} = \frac{3}{100} $$

Assuming "sat" follows three distinct words and there are 100 distinct contexts in total.

##### Combining the Results

$$ P_{KN}(\text{sat} | \text{the cat}) = 0.05 + 0.45 \times (0.125 + 0.525 \times 0.03) $$

##### Final Result

$$ P_{KN}(\text{sat} | \text{the cat}) = 0.05 + 0.45 \times (0.125 + 0.01575) = 0.05 + 0.45 \times 0.14075 = 0.05 + 0.0633375 = 0.1133375 $$


##### Differene with Absolute Discounting

**Back-off Weight**: Kneser-Neys back-off weight $ \lambda(w_{i-n+1}^{i-1}) = \frac{D \cdot | \{ w_i : c(w_{i-n+1}^i) > 0 \} |}{c(w_{i-n+1}^{i-1})} $  depends on the number of distinct words that can follow a context, whereas Absolute Discounting uses a simple normalization factor  </br>
**Distinct Contexts**: Kneser-Ney considers the number of different contexts in which a word appears, while Absolute Discounting does not.

Conclusion : absolute discounting simply subtracts a fixed value from all observed counts, without considering the context in which words appear. Kneser-Ney, on the other hand, adjusts probabilities based on the distribution of contexts, making it more robust and accurate.

#### Base Wav2Vec

```python
from datasets import load_dataset

dataset = load_dataset("hf-internal-testing/librispeech_asr_demo", "clean", split="validation")
dataset
```

We then load an audio sample with its ground-truth transcription.

```python
import IPython.display as ipd

audio_sample = dataset[2]
print(audio_sample["text"].lower())
ipd.Audio(data=audio_sample["audio"]["array"], autoplay=True, rate=audio_sample["audio"]["sampling_rate"])
```

Next, we evaluate vanilla Wav2Vec2 against this single sample.

```python
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC

processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-100h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-100h")
```

We then process the data and pass it to the loaded model to check its output.

```python
inputs = processor(audio_sample["audio"]["array"], sampling_rate=16_000, return_tensors="pt")
```

```python
with torch.no_grad():
  logits = model(**inputs).logits

predicted_ids = torch.argmax(logits, dim=-1)
transcription = processor.batch_decode(predicted_ids)

transcription[0].lower()
```

Comparing the two, we get:

```python
audio_text = audio_sample["text"].lower()
transcription_text = transcription[0].lower()

discrepancies = compare_strings(audio_text, transcription_text)
print(discrepancies)
```
